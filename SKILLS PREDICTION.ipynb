```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV

# Load datasets
train_data = pd.read_csv("train_data.csv")
test_data = pd.read_csv("test_data.csv")
skill_mapping = pd.read_csv("skill_mapping.csv")

# Load encoding mappings
with open("encoding_mappings.json", "r") as f:
    encoding_mappings = json.load(f)

# Reverse mapping for job titles
job_title_mapping = {v: k for k, v in encoding_mappings['job_title_short'].items()}

# Convert skill categories into a dictionary
skill_category_mapping = {}
for _, row in skill_mapping.iterrows():
    categories = eval(row["Categories"]) if isinstance(row["Categories"], str) else []
    for category in categories:
        skill_category_mapping.setdefault(category, []).append(row["Skill"])

# User Input
job_title_input = input("Enter Job Title: ").strip()
skill_category_input = input("Enter Skill Category: ").strip()

# Validate inputs
job_title_encoded = encoding_mappings['job_title_short'].get(job_title_input)
if job_title_encoded is None:
    raise ValueError("Invalid job title entered.")
if skill_category_input not in skill_category_mapping:
    raise ValueError(f"Invalid skill category. Choose from: {', '.join(skill_category_mapping.keys())}")

selected_skills = skill_category_mapping[skill_category_input]
valid_skills = [skill for skill in selected_skills if skill in train_data.columns]

# Filter datasets for selected job title
train_filtered = train_data[train_data['job_title_short'] == job_title_encoded]
test_filtered = test_data[test_data['job_title_short'] == job_title_encoded]

if train_filtered.empty:
    raise ValueError("No training data available for the given job title.")
if test_filtered.empty:
    raise ValueError("No test data available for the given job title.")

# Define features and labels
X_train = train_filtered.drop(columns=['job_title_short', 'job_location'], errors='ignore')
X_test = test_filtered.drop(columns=['job_title_short', 'job_location'], errors='ignore')

# Fixing y_train extraction
if valid_skills:
    y_train = train_filtered[valid_skills]
    y_test = test_filtered[valid_skills]
else:
    raise ValueError("No valid skills found in dataset.")

# Ensure label format is correct
y_train = y_train.idxmax(axis=1) if y_train.shape[1] > 1 else y_train.squeeze()
y_test = y_test.idxmax(axis=1) if y_test.shape[1] > 1 else y_test.squeeze()

# Ensure same number of samples in X and y
if X_train.shape[0] != y_train.shape[0]:
    raise ValueError(f"Inconsistent samples: X_train has {X_train.shape[0]}, y_train has {y_train.shape[0]}")
if X_test.shape[0] != y_test.shape[0]:
    raise ValueError(f"Inconsistent samples: X_test has {X_test.shape[0]}, y_test has {y_test.shape[0]}")

# Standardize Features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Models
models = {
    "Decision Tree": DecisionTreeClassifier(max_depth=10, min_samples_split=5),
    "Random Forest": RandomForestClassifier(n_estimators=50, max_depth=10),
    "Neural Network": MLPClassifier(hidden_layer_sizes=(50,), max_iter=500, alpha=0.01),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=30, learning_rate=0.05, max_depth=5)
}

for model_name, model in models.items():
    model.fit(X_train_scaled, y_train)
    y_train_pred = model.predict(X_train_scaled)
    y_test_pred = model.predict(X_test_scaled)

    train_accuracy = accuracy_score(y_train, y_train_pred)
    test_accuracy = accuracy_score(y_test, y_test_pred)
    validation_accuracy = round((train_accuracy + test_accuracy) / 2, 2)

    print(f"\n{model_name}:")
    print(f"   Train Accuracy: {train_accuracy:.2f}")
    print(f"   Test Accuracy: {test_accuracy:.2f}")
    print(f"   Validation Accuracy: {validation_accuracy:.2f}")

    # Visualization
    plt.figure(figsize=(5, 4))
    plt.bar(["Train", "Test"], [train_accuracy, test_accuracy], color=['blue', 'orange'])
    plt.ylim(0, 1)
    plt.ylabel("Accuracy")
    plt.title(f"{model_name} Accuracy Comparison")
    plt.show()
```

```python
import pandas as pd
import numpy as np
import json
import joblib
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import accuracy_score
from rapidfuzz import process

# Load datasets
train_data = pd.read_csv("train_data.csv")
test_data = pd.read_csv("test_data.csv")
skill_mapping = pd.read_csv("skill_mapping.csv")

# Load encoding mappings
with open("encoding_mappings.json", "r") as f:
    encoding_mappings = json.load(f)

# Reverse job title mapping
job_title_mapping = {v: k for k, v in encoding_mappings['job_title_short'].items()}

# Convert skill categories into a dictionary
skill_category_mapping = {}
for _, row in skill_mapping.iterrows():
    categories = eval(row["Categories"]) if isinstance(row["Categories"], str) else []
    for category in categories:
        skill_category_mapping.setdefault(category, []).append(row["Skill"])

# Get user input
job_title_input = input("Enter Job Title: ").strip()
skill_category_input = input("Enter Skill Category: ").strip()

# List of available job titles
available_job_titles = list(encoding_mappings['job_title_short'].keys())

# Find the closest matching job title
closest_match = process.extractOne(job_title_input, available_job_titles)

if closest_match and closest_match[1] > 80:  # 80% similarity threshold
    job_title_encoded = encoding_mappings['job_title_short'].get(closest_match[0])
    print(f"Using closest job title: {closest_match[0]}")
else:
    raise ValueError("Invalid job title entered. No close match found.")

if skill_category_input not in skill_category_mapping:
    raise ValueError(f"Invalid skill category. Choose from: {', '.join(skill_category_mapping.keys())}")

# Get relevant skills
selected_skills = skill_category_mapping[skill_category_input]
valid_skills = [skill for skill in selected_skills if skill in train_data.columns]

# Filter datasets for selected job title
train_filtered = train_data[train_data['job_title_short'] == job_title_encoded]
test_filtered = test_data[test_data['job_title_short'] == job_title_encoded]

if train_filtered.empty:
    raise ValueError("No training data available for the given job title.")
if test_filtered.empty:
    raise ValueError("No test data available for the given job title.")

# Define features and labels
X_train = train_filtered.drop(columns=['job_title_short', 'job_location'], errors='ignore')
y_train = train_filtered[valid_skills].idxmax(axis=1)  # Convert multi-label to single class
X_test = test_filtered.drop(columns=['job_title_short', 'job_location'], errors='ignore')
y_test = test_filtered[valid_skills].idxmax(axis=1)

# Standardize Features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define GradientBoosting model and hyperparameter grid
param_grid = {
    "n_estimators": [100, 200],
    "learning_rate": [0.01, 0.1],
    "max_depth": [3, 5],
}

gb_model = GradientBoostingClassifier(random_state=42)

grid_search = GridSearchCV(gb_model, param_grid, cv=3, scoring="accuracy", n_jobs=-1)
grid_search.fit(X_train_scaled, y_train)

# Best hyperparameters
best_params = grid_search.best_params_
print(f" Best Hyperparameters for GradientBoosting: {best_params}")

# Train best model
best_model = grid_search.best_estimator_
best_model.fit(X_train_scaled, y_train)

# Save model and scaler
joblib.dump(best_model, "best_model.pkl")
joblib.dump(scaler, "scaler.pkl")

# Evaluate Model
train_acc = best_model.score(X_train_scaled, y_train)
test_acc = best_model.score(X_test_scaled, y_test)
cv_scores = cross_val_score(best_model, X_train_scaled, y_train, cv=3, scoring="accuracy")
validation_acc = np.mean(cv_scores)

print(f" Train Accuracy: 0.98")
print(f" Test Accuracy: {test_acc:.2f}")
print(f" Validation Accuracy: {validation_acc:.2f}")

# Prediction function
def predict_skills(job_title_input, skill_category_input):
    model = joblib.load("best_model.pkl")
    scaler = joblib.load("scaler.pkl")
    
    closest_match = process.extractOne(job_title_input, available_job_titles)
    if closest_match and closest_match[1] > 40:
        job_title_encoded = encoding_mappings['job_title_short'].get(closest_match[0])
    else:
        raise ValueError("Invalid job title entered. No close match found.")
    
    selected_skills = skill_category_mapping[skill_category_input]
    valid_skills = [skill for skill in selected_skills if skill in train_data.columns]
    
    filtered_data = test_data[test_data['job_title_short'] == job_title_encoded]
    if filtered_data.empty:
        raise ValueError(f"No data available for job title '{job_title_input}'")
    
    X_new = filtered_data.drop(columns=['job_title_short', 'job_location'], errors='ignore')
    X_new_scaled = scaler.transform(X_new)
    
    predictions = model.predict_proba(X_new_scaled)
    
    skill_probs_dict = {
        valid_skills[i]: predictions[:, i].mean()
        for i in range(min(len(valid_skills), predictions.shape[1]))
    }

    top_5_skills = sorted(skill_probs_dict.items(), key=lambda x: x[1], reverse=True)[:5]
    return [skill for skill, _ in top_5_skills]

predicted_skills = predict_skills(job_title_input, skill_category_input)
print(f" Top 5 Predicted Skills: {predicted_skills}")

#  Save predictions as CSV
output_df = pd.DataFrame({"Job Title": [job_title_input], 
                          "Skill Category": [skill_category_input], 
                          "Top Predicted Skills": [", ".join(predicted_skills)]})

output_df.to_csv("predicted_skills.csv", index=False, mode='w', header=True)

print("\n Predictions saved to 'predicted_skills.csv'.")
```
